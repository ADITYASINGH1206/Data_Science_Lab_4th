```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

Subplot in matplotlib

```python
plt.subplot(2,2,1)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.subplot(2,2,4)
```

Object Oriented way

```python
fig=plt.figure(figsize=(10,5),dpi=100)
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
plt.title("Main Plot")
axis2=fig.add_axes([0.3,0.3,0.5,0.5])
axis2.plot(y,x)
plt.title("Inset Plot")
```

Types of plots:

```python
plt.scatter(x,y)
```

```python
plt.hist(y,bins=5)
```

```python
plt.boxplot(y)
```

Saving plot

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("myplot.png")
```

Working with images

```python
import matplotlib.image as mpimg
img=mpimg.imread("myplot.png")
plt.imshow(img)
plt.axis('off')
plt.show()
```

Optional Desingning

```python
plt.plot(x,y,color='purple',linewidth=3,alpha=0.5,linestyle='--',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
```
