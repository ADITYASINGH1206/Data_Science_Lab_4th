#  Seaborn – Introduction & Relational Plots


## What is Seaborn?

Seaborn is a **high-level data visualization library** built on top of Matplotlib.  
It provides:
- Beautiful, modern default styles  
- Concise syntax (less code for complex visuals)  
- Integration with Pandas DataFrames  

Perfect for quick, attractive plots while keeping full Matplotlib power underneath.


###  What Are Relational Plots?

Relational plots are used to **show the relationship between two variables**.

- They focus on how one variable changes with respect to another.
- Typically used for continuous (numeric) data, but can also mix with categories.

In Seaborn, the main relational plots are:

1. **`scatterplot()`** → displays individual data points  
2. **`lineplot()`** → shows a continuous trend or summary line

> Both functions belong to Seaborn’s *relational* family.


## You can load sample DataSets using SNS

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Built-in sample datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
penguins = sns.load_dataset("penguins")

print(tips.head())
print(iris.head())
print(penguins.head())
```

### Scatter Plot with scatterplot()

```python
# Scatter: Total bill vs Tip
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

### Line Plot with lineplot()

```python
# Line plot: day vs tip (average per day)
sns.lineplot(x="day", y="tip", data=tips, estimator="mean", marker="o")
plt.title("Line Plot: Average Tip by Day")
plt.show()
```

### Styling & Color Palettes

```python
# Change global style
sns.set_style("darkgrid")        # options: white, dark, darkgrid, whitegrid, ticks

# Try different color palettes
sns.set_palette("pastel")         # others: deep, bright, dark, muted, colorblind

sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips, s=80)
plt.title("Styled Scatter with Hue & Palette")
plt.show()
```

# Seaborn – Categorical Plots


## What Are Categorical Plots?

Categorical plots visualize data where at least one variable is **categorical**  
(e.g., day of week, gender, class).

Seaborn gives several easy options to:
- Compare **counts** or **averages** across categories
- Understand spread & distribution


## barplot() vs countplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Barplot: shows average value of a numeric variable per category
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Barplot: Avg Total Bill by Day")
plt.show()

# Countplot: shows frequency of each category
sns.countplot(x="day", data=tips)
plt.title("Countplot: Number of Records by Day")
plt.show()
```

## boxplot() & violinplot()

```python
# Boxplot: spread, median & outliers
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Boxplot: Total Bill by Day")
plt.show()

# Violinplot: combines boxplot + distribution
sns.violinplot(x="day", y="total_bill", data=tips, inner="box")
plt.title("Violinplot: Total Bill by Day")
plt.show()
```

## stripplot() & swarmplot()

```python
# Stripplot: shows all points (can overlap)
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("Stripplot: Total Bill by Day")
plt.show()

# Swarmplot: adjusts points so they don’t overlap
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("Swarmplot: Total Bill by Day")
plt.show()
```

| Plot        | Best for |
|-------------|-----------|
| **barplot** | Compare average values across categories |
| **countplot** | Show how many items fall in each category |
| **boxplot** | See median, quartiles, and outliers |
| **violinplot** | Same as boxplot + shows distribution |
| **stripplot** | View every observation (good for small data) |
| **swarmplot** | Like stripplot but avoids overlap |


# Seaborn – Distribution Plots

## What Are Distribution Plots?

Distribution plots show how data values are **spread or concentrated**.

Seaborn offers:
- **`histplot()`** → modern histogram (with optional KDE)
- **`kdeplot()`** → smooth curve of distribution
- **`rugplot()`** → small marks showing individual values  
*(Older `distplot()` is now replaced by `histplot()`.)*


## histplot() – Basic Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Basic histogram
sns.histplot(tips["total_bill"], bins=20, color="skyblue")
plt.title("Histogram of Total Bill")
plt.show()
```

## kdeplot() – Kernel Density Estimate

```python
# Smooth curve of the distribution
sns.kdeplot(tips["total_bill"], fill=True, color="green")
plt.title("KDE Plot of Total Bill")
plt.show()
```

## distplot() (Legacy) vs histplot()

> `distplot()` is deprecated.
> Use `histplot()` or `kdeplot()` instead.


## rugplot() – Small Ticks for Observations

```python
sns.rugplot(x=tips["total_bill"], height=0.1, color="red")
plt.title("Rugplot: Total Bill Values")
plt.show()
```

## Multiple Distributions with hue

```python
sns.histplot(x="total_bill", hue="sex", data=tips, kde=True, bins=20, palette="Set2")
plt.title("Histogram by Gender (with KDE)")
plt.show()
```

Key takeaways

histplot() → flexible histogram (can combine with KDE).

kdeplot() → smooth density curve.

rugplot() → quick look at raw values.

distplot() is deprecated → prefer histplot().

Use hue to compare groups in the same distribution plot.

# Seaborn – Class 4: Regression & Mixed Plots

## Regression & Mixed Plots

Seaborn provides plots that combine **relationships** and **statistical models** or mix multiple plot types.

- `regplot()` / `lmplot()` → scatter + regression line  
- `jointplot()` → scatter with hist/KDE  
- `pairplot()` → multiple relationships in one go


# regplot() – Scatter + Regression Line

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Scatter with regression line
sns.regplot(x="total_bill", y="tip", data=tips, color="green")
plt.title("regplot(): Bill vs Tip with Regression Line")
plt.show()
```

# lmplot() – Regression with Facets

```python
# lmplot is similar but creates its own figure and supports grouping
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips, height=5, aspect=1.2)
plt.title("lmplot(): Regression by Gender")
plt.show()
```

# jointplot() – Scatter + Histogram / KDE

```python
# Scatter with marginal histograms
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")

# Or with KDE contours
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde", fill=True)
```

# pairplot() – Relationships Between Many Variables

```python
# Pairwise relationships for multiple columns
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
plt.suptitle("Pairplot: Iris Dataset", y=1.02)
plt.show()
```

#  Seaborn: Matrix/Grid Plots + Styling 


## Why Matrix / Grid Plots?

Matrix or grid-style plots are perfect for:
- Showing correlations between variables  
- Visualizing pivot tables  
- Discovering patterns in large datasets


### Heatmaps (heatmap())

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Correlation heatmap
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Cluster Map (clustermap())

```python
# Clustered heatmap
sns.clustermap(corr, annot=True, cmap="viridis")
```

### Styling & Themes

```python
# Change global style
sns.set_style("darkgrid")       # options: white, dark, darkgrid, whitegrid, ticks

# Change color palette
sns.set_palette("Set2")

# Apply to a plot
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips, s=80)
plt.title("Styled Scatter with Darkgrid & Palette")
plt.show()
```
